## Method 1: Direct Connection

## Method 2: With Conjugate

```python
import pandas as pd;from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://myla:Myla123*@localhost:3306/rritec')
with engine.connect() as conn:
    rs =conn.execute("select * from emp")
    df=pd.DataFrame(data=rs.fetchall(),columns=rs.keys())
df
```

## Method 3: Using Pandas

```python
import pandas as pd;from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://myla:Myla123*@localhost:3306/rritec')
conn=engine.connect()
pd.read_sql("select * from emp",conn)
```
