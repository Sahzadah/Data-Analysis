## Method 2: With Conjugate

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://myla:Myla123*@localhost:3306/test')
with engine.connect() as conn:
    rs =conn.execute("select * from product_detail pd")
    df=pd.DataFrame(data=rs.fetchall(),columns=rs.keys())
df
```
